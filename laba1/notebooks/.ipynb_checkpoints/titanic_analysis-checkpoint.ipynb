```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt
import seaborn as sns

print("Библиотеки успешно загружены")
```

```python
# Загрузка данных
train_df = pd.read_csv('../data/train.csv')
test_df = pd.read_csv('../data/test.csv')

print("Данные успешно загружены")
print(f"Размер train данных: {train_df.shape}")
print(f"Размер test данных: {test_df.shape}")
print(f"\nКолонки train: {list(train_df.columns)}")
print(f"Колонки test: {list(test_df.columns)}")
```

```python
# Первые 5 строк
print("TRAIN.CSV - ПЕРВЫЕ 5 СТРОК:")
display(train_df.head())

print("TEST.CSV - ПЕРВЫЕ 5 СТРОК:")
display(test_df.head())
```

```python
# Анализ пропущенных значений в train
print("ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ В TRAIN.CSV:")
missing_train = train_df.isnull().sum()
print(missing_train[missing_train > 0])

# Анализ пропущенных значений в test
print("ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ В TEST.CSV:")
missing_test = test_df.isnull().sum()
print(missing_test[missing_test > 0])
```

```python
# Создаем копии данных
train_filled = train_df.copy()
test_filled = test_df.copy()

# Заполняем пропуски в TRAIN данных
print("Заполнение пропусков в TRAIN данных:")
for col in train_filled.columns:
    if train_filled[col].isnull().sum() > 0:
        if train_filled[col].dtype == 'object':  # Категориальные признаки
            # Заполняем модой (самое частое значение)
            cabin_mode = train_filled[col].mode()[0]
            train_filled[col].fillna(cabin_mode, inplace=True)
            print(f"  {col}: заполнен модой ({cabin_mode})")
        else:  # Числовые признаки
            # Заполняем медианой
            age_median = train_filled[col].median()
            train_filled[col].fillna(age_median, inplace=True)
            print(f"  {col}: заполнен медианой ({age_median:.2f})")

print("\nЗаполнение пропусков в TEST данных:")
# Заполняем пропуски в TEST данных
for col in test_filled.columns:
    if test_filled[col].isnull().sum() > 0:
        if test_filled[col].dtype == 'object':  # Категориальные признаки
            # Заполняем модой (самое частое значение)
            cabin_mode = test_filled[col].mode()[0]
            test_filled[col].fillna(cabin_mode, inplace=True)
            print(f"  {col}: заполнен модой ({cabin_mode})")
        else:  # Числовые признаки
            # Заполняем медианой
            age_median = test_filled[col].median()
            test_filled[col].fillna(age_median, inplace=True)
            print(f"  {col}: заполнен медианой ({age_median:.2f})")

# Показываем, что пропуски были заполнены
print("\nРезультат заполнения:")
print("Пропуски в TRAIN после заполнения:", train_filled.isnull().sum().sum())
print("Пропуски в TEST после заполнения:", test_filled.isnull().sum().sum())
```

```python
# Пункт 5: Нормализация данных

from sklearn.preprocessing import StandardScaler

# Выбираем числовые столбцы для нормализации
numeric_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
numeric_cols = [col for col in numeric_cols if col in train_filled.columns]

print("Числовые столбцы для нормализации:", numeric_cols)

standard_scaler = StandardScaler()  # Z-оценка (стандартизация)

print("\nНормализация TRAIN данных:")
# StandardScaler
train_standard = train_filled.copy()
train_standard[numeric_cols] = standard_scaler.fit_transform(train_filled[numeric_cols])
print("StandardScaler применен к TRAIN")

print("\nНормализация TEST данных:")
# StandardScaler
standard_scaler_test = StandardScaler()
standard_scaler_test.fit(train_filled[numeric_cols])  # Обучаем на TRAIN данных
test_standard = test_filled.copy()
test_standard[numeric_cols] = standard_scaler_test.transform(test_filled[numeric_cols])
print("StandardScaler применен к TEST")

# Показываем результаты нормализации
print("\n" + "="*50)
print("РЕЗУЛЬТАТЫ НОРМАЛИЗАЦИИ")
print("="*50)

print("\nTRAIN данные после StandardScaler (Z-оценка):")
print(train_standard[numeric_cols].describe().round(3))

print("\nПроверка: Среднее ≈ 0, Стандартное отклонение ≈ 1")
print("StandardScaler - Средние значения:")
print(train_standard[numeric_cols].mean().round(3))

print("\nStandardScaler - Стандартные отклонения:")
print(train_standard[numeric_cols].std().round(3))

# Демонстрация первых строк
print("\n" + "="*50)
print("ПЕРВЫЕ СТРОКИ ПОСЛЕ НОРМАЛИЗАЦИИ")
print("="*50)

print("\nStandardScaler (первые 3 строки):")
display(train_standard[numeric_cols].head(3))
```

```python
# Выбираем категориальные столбцы для преобразования
categorical_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
categorical_cols = [col for col in categorical_cols if col in train_filled.columns]

print("Категориальные столбцы для преобразования:", categorical_cols)

# Показываем уникальные значения до преобразования
print("\nУникальные значения до OHE:")
for col in categorical_cols:
    print(f"{col}: {train_filled[col].unique()}")

# Применяем One-Hot Encoding с помощью pd.get_dummies()
print("\nПрименение One-Hot Encoding...")

# Для TRAIN данных
train_ohe = pd.get_dummies(train_filled, columns=categorical_cols, drop_first=True)
print("✓ OHE применен к TRAIN данным")

# Для TEST данных (важно использовать те же колонки!)
test_ohe = pd.get_dummies(test_filled, columns=categorical_cols, drop_first=True)
print("✓ OHE применен к TEST данным")

# Выравниваем колонки в test с train (чтобы избежать переобучения)
print("\nВыравнивание колонок TEST с TRAIN...")

# Находим колонки, которые есть в train но нет в test
missing_cols = set(train_ohe.columns) - set(test_ohe.columns)

# Добавляем недостающие колонки в test с значениями 0
for col in missing_cols:
    test_ohe[col] = 0

# Убеждаемся что колонки в том же порядке
test_ohe = test_ohe[train_ohe.columns]
print("✓ Колонки выравнены")

# Показываем результаты преобразования
print("\n" + "="*60)
print("РЕЗУЛЬТАТЫ ONE-HOT ENCODING")
print("="*60)

print(f"Размер TRAIN до OHE: {train_filled.shape}")
print(f"Размер TRAIN после OHE: {train_ohe.shape}")

print(f"\nРазмер TEST до OHE: {test_filled.shape}")
print(f"Размер TEST после OHE: {test_ohe.shape}")

print("\nНовые бинарные колонки (первые 10):")
new_columns = [col for col in train_ohe.columns if any(col.startswith(cat) for cat in categorical_cols)]
print(new_columns[:10])

# Демонстрация преобразованных данных
print("\n" + "="*60)
print("ПРИМЕР ПРЕОБРАЗОВАННЫХ ДАННЫХ")
print("="*60)

print("TRAIN - первые 3 строки после OHE:")
display(train_ohe.head(3))

print("\nКатегориальные колонки преобразованы в бинарные:")
for col in categorical_cols:
    ohe_cols = [c for c in train_ohe.columns if c.startswith(col)]
    if ohe_cols:
        print(f"{col} → {ohe_cols}")
```
